

## Dependencies and starter code

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np

# Study data files
mouse_metadata = "Resources/Mouse_metadata.txt"
study_results = "Resources/Study_results.txt"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)

# Combine the data into a single dataset
combined_df = pd.merge(mouse_metadata, study_results,how='outer',on='Mouse ID')
print(len(combined_df))
combined_df.head(20)
```

## Summary statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# group dataframe by Drug Regimen
drug_group_df = combined_df.groupby("Drug Regimen")

# find the mean of each regimen
drug_means = drug_group_df["Tumor Volume (mm3)"].mean()

# find median
drug_median = drug_group_df["Tumor Volume (mm3)"].median()

# find variance
drug_variance = drug_group_df["Tumor Volume (mm3)"].var()

# standard deviation
drug_stanDev = drug_group_df["Tumor Volume (mm3)"].std()

# SEM
drug_sem = drug_group_df["Tumor Volume (mm3)"].sem()

# create and display table
drug_summary = pd.DataFrame({"Mean": drug_means, "Median": drug_median, "Variance": drug_variance,
                             "Standard Deviation": drug_stanDev, "SEM": drug_sem})
drug_summary
```

## Bar plots

```python
# Generate a bar plot showing number of data points for each treatment regimen using pandas

# create variable to plot drug regimen data
drug_data = combined_df["Drug Regimen"].value_counts()

# pandas bar chart
drug_chart = drug_data.plot(kind = "bar", title = "Data by Drug Regimen")
drug_chart.set_xlabel("Drug Regimen")
drug_chart.set_ylabel("Data Points")
```

```python
# Generate a bar plot showing number of data points for each treatment regimen using pyplot

# create simple variable for x-axis -- y-axis will be drug_data
drug_name = np.arange(len(drug_group_df))

# pyplot bar chart
plt.bar(drug_name, drug_data, color = 'blue', alpha = .75, align = "center")
tick_locations = [value for value in drug_name]
plt.xticks(tick_locations,["Capomulin","Ceftamin","Infubinol","Ketapril","Naftisol","Placebo",
                           "Proriva","Ramicane","Stelasyn", "Zoniferol"],rotation="vertical")

# labels / title and display
plt.title("Data by Drug Regimen")
plt.xlabel("Drug Regimen")
plt.ylabel("Data Points")
plt.show()
plt.tight_layout()
```

## Quartiles, outliers and boxplots

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. Calculate the IQR and quantitatively determine if there are any potential outliers.
# Capomulin, Ramicane, Infubinol, Ceftamin

# separate drug regimens by pulling 4 most promising
good_treatment = combined_df.loc[(combined_df["Drug Regimen"] == "Capomulin") |
                                 (combined_df["Drug Regimen"] == "Ramicane") |
                                 (combined_df["Drug Regimen"] == "Infubinol") |
                                 (combined_df["Drug Regimen"] == "Ceftamin")]

# find the final tumor volume for each mouse
final_tumor = good_treatment.loc[(good_treatment["Timepoint"] == 45),:]
final_tumor.head()
```

```python
# list and iqr for Capomulin
cap_df = final_tumor.loc[(final_tumor["Drug Regimen"] == "Capomulin"),:]
cap_tumor = cap_df["Tumor Volume (mm3)"]
cap_quant = cap_tumor.quantile([.25,.5,.75])
lowercap = cap_quant[.25]
uppercap = cap_quant[.75]
cap_iqr = uppercap-lowercap

# potential outliers
cap_lowerbound = lowercap - (1.5*cap_iqr)
cap_upperbound = uppercap + (1.5*cap_iqr)

# print
print(f"Lower quartile: {lowercap} mm3.")
print(f"Upper quartile: {uppercap} mm3.")
print(f"IQR: {cap_iqr}")
print(f"Values below {cap_lowerbound} could be outliers.")
print(f"Values above {cap_upperbound} could be outliers.")
```

```python
# list for Ramicane
ram_df = final_tumor.loc[(final_tumor["Drug Regimen"] == "Ramicane"),:]
ram_tumor = ram_df["Tumor Volume (mm3)"]
ram_quant = ram_tumor.quantile([.25,.5,.75])
lowerram = ram_quant[.25]
upperram = ram_quant[.75]
ram_iqr = upperram-lowerram

# potential outliers
ram_lowerbound = lowerram - (1.5*ram_iqr)
ram_upperbound = upperram + (1.5*ram_iqr)

# print
print(f"Lower quartile: {lowerram} mm3.")
print(f"Upper quartile: {upperram} mm3.")
print(f"IQR: {ram_iqr}")
print(f"Values below {ram_lowerbound} could be outliers.")
print(f"Values above {ram_upperbound} could be outliers.")
```

```python
# list for Infubinol
inf_df = final_tumor.loc[(final_tumor["Drug Regimen"] == "Infubinol"),:]
inf_tumor = inf_df["Tumor Volume (mm3)"]
inf_quant = inf_tumor.quantile([.25,.5,.75])
lowerinf = inf_quant[.25]
upperinf = inf_quant[.75]
inf_iqr = upperinf - lowerinf

# potential outliers
inf_lowerbound = lowerinf - (1.5*inf_iqr)
inf_upperbound = upperinf + (1.5*inf_iqr)

# print
print(f"Lower quartile: {lowerinf} mm3.")
print(f"Upper quartile: {upperinf} mm3.")
print(f"IQR: {inf_iqr}")
print(f"Values below {inf_lowerbound} could be outliers.")
print(f"Values above {inf_upperbound} could be outliers.")
```

```python
# list for Ceftamin
cef_df = final_tumor.loc[(final_tumor["Drug Regimen"] == "Ceftamin"),:]
cef_tumor = cef_df["Tumor Volume (mm3)"]
cef_quant = cef_tumor.quantile([.25,.5,.75])
lowercef = cef_quant[.25]
uppercef = cef_quant[.75]
cef_iqr = uppercef-lowercef

# potential outliers
cef_lowerbound = lowercef - (1.5*cef_iqr)
cef_upperbound = uppercef + (1.5*cef_iqr)

# print
print(f"Lower quartile: {lowercef} mm3.")
print(f"Upper quartile: {uppercef} mm3.")
print(f"IQR: {cef_iqr}")
print(f"Values below {cef_lowerbound} could be outliers.")
print(f"Values above {cef_upperbound} could be outliers.")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
all_tumor = [cap_tumor, ram_tumor, inf_tumor, cef_tumor]
label = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
fig1, ax1 = plt.subplots()
ax1.set_title("Final Tumors")
ax1.set_xlabel("Drug Regimen")
ax1.set_ylabel("Tumor Size mm3")
ax1.boxplot(all_tumor, labels = label)
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen

# create dataframe of Capomulin average tumor volume
capomulin_df = combined_df.loc[(combined_df["Drug Regimen"] == "Capomulin"),:]

cap_avg_df = pd.DataFrame(capomulin_df.groupby(["Mouse ID", "Weight (g)"])["Tumor Volume (mm3)"].mean()).reset_index()

cap_avg_df = cap_avg_df.rename(columns={"Tumor Volume (mm3)": "Average Tumor Volume (mm3)"})
cap_avg_df
```

```python
# Calculate the correlation coefficient and linear regression model for mouse weight and average tumor volume for the Capomulin regimen

# correlation coefficient
x_values = cap_avg_df["Weight (g)"]
y_values = cap_avg_df["Average Tumor Volume (mm3)"]
correlation = st.pearsonr(x_values,y_values)
print(f"The correlation between weight and average tumor volume is {round(correlation[0],3)}.")

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)

regress_values = x_values * slope + intercept

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(20,36),fontsize=15,color="red")    # show the equation in plot
plt.title("Average Tumor Volume by Mouse Weight for Capomulin")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```
